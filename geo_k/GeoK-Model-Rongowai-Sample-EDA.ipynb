```python
# imports

import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load .npy files

compressed_ddms = list()
path = "..\\..\\data\\Rongowai\\GEO-K_exampledata\\"

for file in os.listdir(path):
    if file.startswith("compressed_sample_") and file.endswith(".npy"):
        compressed_ddms.append(np.load(os.path.join(path, file)))
```

```python
# plot all the compressed DDMs

plt.figure(figsize=(12, 4))
for i, d in enumerate(compressed_ddms):
    plt.plot(d, label=f"Compressed DDM {i+1}")
plt.title("Compressed DDMs")
plt.xticks(ticks=np.arange(len(compressed_ddms[0])), labels=np.arange(1,len(compressed_ddms[0])+1), rotation=0)
plt.xlabel("Sample Index")
plt.ylabel("Compressed values")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
